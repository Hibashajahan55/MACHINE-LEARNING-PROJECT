```python

# skewness correction
# Features that need log transformation
highly_skewed_features = ['n_non_stop_words','n_unique_tokens','n_non_stop_unique_tokens','kw_max_min','shares','kw_avg_min','self_reference_min_shares',
'self_reference_avg_sharess','kw_max_avg','self_reference_max_shares','kw_min_max','num_videos','kw_avg_avg','num_self_hrefs','num_hrefs']

# Apply log1p transformation
for col in highly_skewed_features:
    df[col + '_log'] = np.log1p(df[col])

# Recalculate skewness after transformation
print(df[[col + '_log' for col in highly_skewed_features]].skew())

```

```python
# Fix  kw_avg_min_log :NaN Before Log Transformation

# Replace zeros or negative values with a small positive number before log transformation
df['kw_avg_min'] = df['kw_avg_min'].apply(lambda x: x if x > 0 else 1e-5)

# Apply log1p transformation again
df['kw_avg_min_log'] = np.log1p(df['kw_avg_min'])

# Check if NaNs still exist
print(df['kw_avg_min_log'].isna().sum())  # Should be 0 if fixed correctly
```

```python
# Compute skewness for transformed columns
new_skewness_values = df[[col for col in df.columns if '_log' in col]].skew()

# Display the new skewness values sorted
print(new_skewness_values.sort_values(ascending=False))
```

```python
# detecting outliers in the log-transformed features

import matplotlib.pyplot as plt
import seaborn as sns

# Select transformed columns (ending in '_log')
transformed_cols = [col for col in df.columns if '_log' in col]

# Plot boxplots for transformed columns
plt.figure(figsize=(15, 8))
df[transformed_cols].boxplot(rot=90)  # Rotate labels for better visibility
plt.title("Boxplot of Log-Transformed Features to Detect Outliers")
plt.show()
```

```python
# Box Plot for visualize outliers
columns = df.select_dtypes(include = ['number'])

for col in transformed_cols:
    plt.figure(figsize=(10,6))
    columns.boxplot(column = col)
    plt.title(f" Box Plot for {col}")
    plt.ylabel('Values')
    plt.show()

```
