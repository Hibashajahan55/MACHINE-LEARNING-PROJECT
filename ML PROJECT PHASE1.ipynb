# 1.INTRODUCTION

The rapid expansion of the internet and digital media has revolutionized the way news is consumed and shared. With social media playing a crucial role in content dissemination, predicting the popularity of online news articles has become a significant research area. This project focuses on building a machine learning model to predict the number of shares an article is likely to receive, based on a diverse set of features extracted from news articles.

# 2. GOAL OF THE PROJECT

The goal of this project is to develop a predictive model that can estimate the popularity of news articles published on Mashable. Popularity is determined by the number of shares an article receives on social media platforms. By leveraging machine learning techniques, we aim to identify key factors influencing article shares and improve content strategies for online publishers.

Type: Multivariate

Instances: 39,797

Features: 61(58 predictive attributes)

Associated Tasks: Classification, Regression

Feature Types: Integer, Real-valued

### KEY FEATURES OF DATASET

The dataset consists of a variety of attributes that can help in predicting the popularity of news articles, including:

* Content-Based Features: Word count, uniqueness of words, and presence of keywords.
* Media-Related Features: Number of images, videos, and links in the article.
* Publication Timing: Day of the week and whether it was published on a weekend.
* Sentiment and Subjectivity Analysis: Polarity and subjectivity of the article’s content and title.
* Engagement Metrics: Number of shares of previously referenced articles.
* The target variable in this dataset is “shares”, representing the number of times an article was shared on social media.

```python
# Visualizing Skewness using Histograms
skewed_features = skewness[skewness.abs() > 1].index  # Selecting highly skewed features
num_features = len(skewed_features)
num_rows = (num_features // 5) + 1  # Adjust rows dynamically

plt.figure(figsize=(15, num_rows * 3))
for i, feature in enumerate(skewed_features, 1):
    plt.subplot(num_rows, 5, i)
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f'Skewness: {round(skewness[feature], 2)}')
    plt.xlabel(feature)
plt.tight_layout()
plt.show()
```

```python
#skeweness correction
#  Apply Log Transformation for strictly positive features
X_log_transformed = X.copy()
positive_features = [col for col in X.columns if (X[col] > 0).all()]
X_log_transformed[positive_features] = np.log1p(X_log_transformed[positive_features])
```

```python
# Apply Power Transformation to reduce skewness
pt = PowerTransformer(method='yeo-johnson')
X_transformed = pt.fit_transform(X_log_transformed)
X_transformed_df = pd.DataFrame(X_transformed, columns=X.columns)
```

```python
# Rechecking skewness after transformation
skewness_final = X_transformed_df.skew()
print("\nSkewness after Transformation:")
print(skewness_final.sort_values(ascending=False))
```

```python
#Outlier Visualization (Before Handling)
num_features = len(X_transformed_df.columns)
num_rows = (num_features // 3) + (num_features % 3 > 0)

fig, axes = plt.subplots(num_rows, 3, figsize=(18, num_rows * 4))
axes = axes.flatten()

for i, feature in enumerate(X_transformed_df.columns):
    sns.boxplot(x=X_transformed_df[feature], ax=axes[i])
    axes[i].set_title(f"{feature} (Before Outlier Handling)")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

#  Outlier Handling using IQR (AFTER Skewness Correction)
Q1 = X_transformed_df.quantile(0.25)
Q3 = X_transformed_df.quantile(0.75)
IQR = Q3 - Q1
outlier_threshold = 1.5 * IQR

# Capping outliers using Pandas clip function
X_capped_df = X_transformed_df.clip(lower=Q1 - outlier_threshold, upper=Q3 + outlier_threshold, axis=1)

```

```python
# Final Boxplot visualization
num_features = len(X_capped_df.columns)
num_rows = (num_features // 3) + (num_features % 3 > 0)

fig, axes = plt.subplots(num_rows, 3, figsize=(18, num_rows * 4))
axes = axes.flatten()

for i, feature in enumerate(X_capped_df.columns):
    sns.boxplot(x=X_capped_df[feature], ax=axes[i])
    axes[i].set_title(feature)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Checking Skewness Again After Outlier Handling
skewness_final_after_outlier = X_capped_df.skew()
print("\nSkewness after Outlier Handling:")
print(skewness_final_after_outlier.sort_values(ascending=False))
```
